```python
import numpy as np
# a helper function for writing predictions to a file
def write_to_submission_file(predicted_labels, out_file,
                             target='target', index_label="session_id"):
    predicted_df = pd.DataFrame(predicted_labels,
                                index = np.arange(1, predicted_labels.shape[0] + 1),
                                columns=[target])
    predicted_df.to_csv(out_file, index_label=index_label)
```

#### Missing Data

```python
total = df.isnull().sum().sort_values(ascending=False)
percent =  (df.isnull().sum() / df.shape[0]).sort_values(ascending=False)
missing_data = pd.concat([total,percent], 
                         axis=1, 
                         keys=['Total','Percent'])
missing_data.head(20)
```

```python
df1 = df.drop(labels=(missing_data[missing_data['Total'] > 1].index),
              axis=1)
df1 = df1.drop(df1.loc[df1['Electrical'].isnull()].index, 
               axis=0)
df1.isnull().sum().max()
```
