```python
ytrn = dfNum['SalePrice']
```

```python
xtrn = dfNum.drop(columns=['Id','SalePrice','GarageYrBlt','MasVnrArea'],axis=1)
xtrn = xtrn.fillna(0)
```

```python
xtrn.info()
```

```python
from sklearn.linear_model import LinearRegression
mdl = LinearRegression()
mdl.fit(xtrn,ytrn)
```

```python
mdl.score(xtrn,ytrn)
```

```python
dftst = pd.read_csv('data/test.csv')
```

```python
dftst.head()
```

```python
dftstNum = dftst.select_dtypes(include=['float64','int64'])
srsId = dftstNum['Id']
xtst = dftstNum.drop(columns=['Id','GarageYrBlt','MasVnrArea'],axis=1)
xtst = xtst.fillna(0)
xtst.info()
```

```python
yprd = mdl.predict(xtst)
```

```python
yprdDf = pd.DataFrame(data=yprd, columns=['SalePrice'])
dfsub = pd.concat([srsId,yprdDf['SalePrice']],axis=1)
dfsub.head()
```

```python
dfsub.to_csv('submissions/submission_01.txt',index=False)
```

```python
import numpy as np
# a helper function for writing predictions to a file
def write_to_submission_file(predicted_labels, out_file,
                             target='target', index_label="session_id"):
    predicted_df = pd.DataFrame(predicted_labels,
                                index = np.arange(1, predicted_labels.shape[0] + 1),
                                columns=[target])
    predicted_df.to_csv(out_file, index_label=index_label)
```
